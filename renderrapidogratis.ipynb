```python
import tensorflow as tf
import time

# Check if GPU is available
if not tf.config.list_physical_devices('GPU'):
    print("No GPU available. Please ensure a GPU runtime is enabled.")
else:
    print("GPU is available.")
    # Define a simple operation to load the GPU
    @tf.function
    def gpu_load_op(x):
        return tf.matmul(x, x)

    # Create a large tensor to work with
    # Increased tensor size to attempt to use more GPU memory
    tensor_size = 30000  # Adjusted size, this will consume significantly more memory
    x = tf.random.normal((tensor_size, tensor_size), dtype=tf.float32)

    print(f"Starting GPU benchmark with tensor size {tensor_size}x{tensor_size}...")

    # Run the operation in a loop to keep the GPU busy
    # Adjust the sleep time to try and target ~80% utilization
    try:
        while True:
            start_time = time.time()
            gpu_load_op(x)
            end_time = time.time()
            # Basic adjustment: sleep less to increase utilization, more to decrease
            # This is a simple approach and may require manual tuning
            computation_time = end_time - start_time
            # Aim for roughly 80% utilization by sleeping for 20% of the computation time
            sleep_time = computation_time * 0.25 # Adjust this factor
            time.sleep(sleep_time)
            # You can monitor GPU utilization and memory usage using the `nvidia-smi` command in another cell
    except KeyboardInterrupt:
        print("Benchmark stopped.")
```
